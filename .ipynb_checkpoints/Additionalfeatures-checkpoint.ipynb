# Additional geospatial features

```python
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
import pcraster as pc
from osgeo import gdal
import rasterio
import rasterio.mask
import numpy as np
from scipy import stats
import seaborn as sns
from dbfread import DBF
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
#Paths
patht="D:/Erasmus/Thesis/"
pathdat=patht+"/data/"

pathshp=pathdat+"/SHP/SHP2/"
pathrast=pathdat+"/Raster/"
pathpick=pathdat+"/Pickle/"
pathfig=patht+"/Figures/"

pathg=r'D:\Erasmus\Thesis\data\resultsCNN/'
```

```python
gw_sel_int=gpd.read_file(pathshp+"/gw_sel_intmod2.shp")
gw_sel_int['buff_1km']=gw_sel_int.buffer(1000)
```

```python
def tifmap(tiff, mapf):
    dataset = gdal.Open(tiff, gdal.GA_ReadOnly)
    driver = gdal.GetDriverByName('PCRaster')
    outdem = driver.CreateCopy( mapf, dataset)
    outdem=None
    
def maptif( mapf, tiff):
    dataset = gdal.Open(mapf, gdal.GA_ReadOnly)
    driver = gdal.GetDriverByName("GTiff")
    outdem = driver.CreateCopy( tiff, dataset)
    outdem=None
```

```python
#Change the dem format to pcraster #run once|
tifmap(pathrast+"/dem/lsdem1km_4647.tif",pathrast+"/dem/lsdem1km_4647_pc.map")
```

```python
dem= pc.readmap(pathrast+"/dem/lsdem1km_4647_pc.map")
```

## Slope

```python
slpfrac=pc.slope(dem)
pc.report(slpfrac,pathrast+"/dem/slope.map") 
maptif(pathrast+"/dem/slope.map",pathrast+"/dem/slope.tiff")
```

```python
slp1=rasterio.open(pathrast+"/dem/slope.tiff")
gw_sel_int['slope_buff']=np.zeros(len(gw_sel_int))
for i in range(len(gw_sel_int)):
    out_mask, out_transform = rasterio.mask.mask(slp1, gw_sel_int['buff_1km'][i:i+1], crop=True)
    meanval=np.mean(out_mask[out_mask!=slp1.meta['nodata']])
    gw_sel_int['slope_buff'][i]=meanval
```

```python
fig, ax = plt.subplots(figsize=(8,5))
cp=sns.regplot(ax=ax,x='slope_buff', y="r2_2", data=gw_sel_int.sort_values(by=['slope_buff']), fit_reg=True)
plt.grid(alpha=0.5)
```

```python
dat=gw_sel_int.sort_values(by=['slope_buff']).dropna(subset=['slope_buff'])
slope, intercept, r_value, p_value, std_err = stats.linregress(dat["slope_buff"],
                                                                   dat["r2_2"])
print(r_value,p_value)
```

## Drainage density

```python
waterways=gpd.read_file(pathshp+"waterwaysND.shp")
proj_coor=4647
waterways=waterways.to_crs(epsg=proj_coor)
```

```python
gw_sel_int['drainaged']=np.zeros(len(gw_sel_int))
for i in range(len(gw_sel_int)):
    clip = gpd.clip(waterways, gw_sel_int['buff_1km'][i:i+1])
    if len(clip)!=0:
        area= 500*np.pi**2
        dd=sum(clip.length.values)/area
        gw_sel_int['drainaged'][i]=dd
    else:
        gw_sel_int['drainaged'][i]=np.nan
```

```python
fig, ax = plt.subplots(figsize=(8,5))
cp=sns.regplot(ax=ax,x='drainaged', y="r2_2", data=gw_sel_int.sort_values(by=['drainaged']), fit_reg=True)
plt.grid(alpha=0.5)
```

```python
dat=gw_sel_int.sort_values(by=['drainaged']).dropna(subset=['drainaged'])
slope, intercept, r_value, p_value, std_err = stats.linregress(dat["drainaged"],
                                                                   dat["NSE_2"])
print(r_value,p_value)
```

##  Topographic wetness index 

```python
slpper=slpfrac*100
slprad= slpper*0.062831853071796
# Compute the flow direction raster from the DEM
flow_dir = pc.lddcreate(dem, 1e31, 1e31,1e31, 1e31)
# Compute the flow accumulation raster
flow_accum = pc.accuflux(flow_dir,1)
#pc.report(flow_dir, pathrast+"/dem/flowacc_4647_pc.map")

TWI= pc.ln((flow_accum*pc.cellarea()) / pc.tan(slprad) )
pc.report(TWI,pathrast+"/dem/twi.map") 
```

```python
maptif(pathrast+"/dem/twi.map",pathrast+"/dem/twi.tiff")
```

```python
twir=rasterio.open(pathrast+"/dem/twi.tiff")
gw_sel_int['twi']=np.zeros(len(gw_sel_int))
for i in range(len(gw_sel_int)):
    out_mask, out_transform = rasterio.mask.mask(twir, gw_sel_int['buff_1km'][i:i+1], crop=True)
    meanval=np.mean(out_mask[out_mask!=twir.meta['nodata']])
    gw_sel_int['twi'][i]=meanval
```

```python
fig, ax = plt.subplots(figsize=(8,5))
cp=sns.regplot(ax=ax,x='twi', y="r2_2", data=gw_sel_int.sort_values(by=['twi']), fit_reg=True)
plt.grid(alpha=0.5)
```

```python
dat=gw_sel_int.sort_values(by=['twi']).dropna(subset=['twi'])
slope, intercept, r_value, p_value, std_err = stats.linregress(dat["twi"],
                                                                   dat["r2_2"])
print(r_value,p_value)
```

## European soil database maps

```python
sgdb=gpd.read_file(pathshp+"/sgdbe4_0.shp")
proj_coor=4647
sgdb=sgdb.to_crs(epsg=proj_coor)
```

```python
table = DBF(pathshp +'smu_ptrdb.dbf', load=True)
dfsmu=pd.DataFrame(table)
```

```python
sgdbv, cliparea=[], []
f=[]
var="HG"
#cat='4W'
cat='M'
for i in range(len(gw_sel_int)):
#for i in range(10):
    clip = gpd.clip(sgdb, gw_sel_int['buff_1km'][i:i+1])
    
    if len(clip)!=0:
        dfaux=dfsmu[dfsmu.SMU.isin(clip.SMU.values)].reset_index()
        ind=dfaux[dfaux[var]==cat].index.values
        prop=clip.geometry.area.values[ind]/(np.pi*(1000**2))
        sgdbv.append(clip.SMU.values)
        cliparea.append(clip.geometry.area.values/10e6)
    else:
        prop=0
        sgdbv.append(np.nan)
        cliparea.append(np.nan)
    #print(dfaux[var])
    #print(prop.sum())
    f.append(prop.sum())
    #gw_sel_int['p_'+var+'_'+cat][i]=prop.sum()
gw_sel_int['p_'+var+'_'+cat]=f        
#gw_sel_int['sgdbcod']=sgdbv
#gw_sel_int['sgdbarea']=cliparea
```

```python
fig, ax = plt.subplots(figsize=(8,5))
cp=sns.regplot(ax=ax,x='p_'+var+'_'+cat, y="r2_2", data=gw_sel_int.sort_values(by=['p_'+var+'_'+cat]), fit_reg=True)
plt.grid(alpha=0.5)
```

```python
dat=gw_sel_int.sort_values(by=['p_'+var+'_'+cat])
dat1=dat.loc[dat['p_'+var+'_'+cat]!=0]
slope, intercept, r_value, p_value, std_err = stats.linregress(dat1['p_'+var+'_'+cat],
                                                                   dat1["r2_2"])
print(r_value,p_value)
```
