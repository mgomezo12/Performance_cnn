```python
#Import libraries
import rasterio
import rasterio.plot
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
import subprocess
import earthpy as et
import earthpy.spatial as es
import earthpy.plot as ep
from matplotlib_scalebar.scalebar import ScaleBar
import matplotlib.patches as patches
```

```python
patht="D:/Erasmus/Thesis/"
pathdat=patht+"/data/"
pathshp=pathdat+"/SHP/SHP2/"
pathrast=pathdat+"/Raster/"
```

```python
dem=rasterio.open(pathrast+"/dem/lsdem1km_4647.tif")
#LS GIS information OSM
waterbodies=gpd.read_file(pathshp+"waterbodiesND.shp") #to add in the background map
waterways=gpd.read_file(pathshp+"waterwaysND.shp")
#Administrative boundaries
germany_states = gpd.read_file(pathshp+"DEU_adm1.shp")
ND=germany_states[germany_states.NAME_1== "Niedersachsen"]
hg=gpd.read_file(pathshp+"/huek500hyr_ply/huek500hyr_ply.shp") #to add in the background map
citiesND=gpd.read_file(pathshp+"citiesND2.shp")
```

```python
proj_coor=4647
waterbodies=waterbodies.to_crs(epsg=proj_coor)
waterways=waterways.to_crs(epsg=proj_coor)
cities=citiesND.to_crs(epsg=proj_coor)
germany_states=germany_states.to_crs(epsg=proj_coor)
ND=ND.to_crs(epsg=proj_coor)
```

```python
hs=rasterio.open(pathrast+"/dem/hillshade.tif")
```

```python
dictionary={
'Nordfriesische Watten, Marschinseln und Halligen': 'North Frisian Wadden sea, marsh islands and halligen',
'Marschen': 'Marshes',
'Nord- und mitteldeutsches Mittelpleistozän': 'North-Central  Middle Pleistocene',
'Niederungen im nord- und mitteldeutschen Lockergesteinsgebiet': 'North-Central lowlands in unconsolidated rock',
'Nordwestdeutsches Bergland': 'Northwest Uplands',
'Subherzyne Senke': 'Subhercynian Trough',
'Münsterländer Kreidebecken': 'Münsterland Chalk Basin',
'Sandmünsterland': 'Sand Münsterland',
'Mitteldeutsches Buntsandstein': 'Central Bunter sandstone',
'Mitteldeutsches Grundgebirge': 'Central crystalline basement',
'Thüringische Senke': 'Thuringian Basin'
}
hg['R_TEXTeng']=hg.R_TEXT.map(dictionary)
```

```python
colors=["#974649","#008200","#a6acd0","#DFC7A7",
        "#9FF8F7","white","#F2C3D6","#A1D197",
        "#FDF8A4","#9E8833","#91AF2F","#E9E15D"]
# create the colormap using ListedColormap
cmap = ListedColormap(colors)

fig, ax = plt.subplots(figsize=(11, 11))
rasterio.plot.show(hs, ax=ax, alpha=1, cmap="gray")
#ax.imshow(hillshade_angle_10)
waterbodies.plot(ax=ax)
gw=hg.plot(column='R_TEXTeng' , categorical=True, cmap=cmap, linewidth=.6, edgecolor='none',
         legend=True, legend_kwds={'bbox_to_anchor':(.3, 1.05),'fontsize':16,'frameon':False,'markerscale': 1.5},
           ax=ax, zorder=0, alpha=0.7)

wb=waterbodies.plot( ax=ax, alpha=0.8, color='b', linewidth=0.8, zorder=3)
ww=waterways.plot( ax=ax, alpha=0.3, color='b', linewidth=.5,zorder=2)
cit=cities.plot( ax=ax, alpha=1, color='k',markersize=30,zorder=3)
NS=ND.boundary.plot( ax=ax, alpha=1, edgecolor='k', linewidth=0.8,zorder=4)

for x, y, label in zip(cities.geometry.x, cities.geometry.y, cities.name):
    if label== "Bremen":
        yp=-25
    else:
        yp=-15
    NS.annotate(label, xy=(x, y), xytext=(-15, yp), textcoords="offset points", fontsize=15, color="k", alpha=0.9)

leg = ax.get_legend()
leg.set_bbox_to_anchor((0., 0., 0.4, 0.4))

#North arrow
startx, endx = gw.get_xlim()
starty, endy = gw.get_ylim()
arrx=endx- endx*0.002
arry=endy-endy*0.0010
gw.text(x=arrx-arrx*0.0001, y=arry, s='N', fontsize=20,alpha=0.8)
gw.arrow(arrx, arry-arry*0.002, 0, 10000, length_includes_head=True,
          head_width=8000, head_length=20000, overhang=.2, ec="k",facecolor='k', alpha=0.7)

#scalebar
scalebar = ScaleBar(0.8, "m", dimension="si-length", length_fraction=0.10, location="lower right")
gw.add_artist(scalebar)
startx, endx = gw.get_xlim()
starty, endy = gw.get_ylim()

gw.spines['top'].set_visible(False)
gw.spines['right'].set_visible(False)
gw.spines['bottom'].set_visible(False)
gw.spines['left'].set_visible(False)
gw.get_xaxis().set_ticks([])
gw.get_yaxis().set_ticks([])
plt.subplots_adjust(bottom=0.2)

pathfig="D:\FOSTER\Figs/" 
#plt.savefig(pathfig+"hgeng.jpg",bbox_inches="tight",dpi=250)
```
